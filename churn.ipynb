```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
churn = pd.read_csv('churn.txt')
churn.head()
```

```python
fig, ax = plt.subplots(figsize=(7,7))
sns.countplot(x="Churn?", data=churn,ax=ax)
plt.savefig("churn.png")
```

```python
fig, ax = plt.subplots(figsize=(7,7))

ip_ct=pd.crosstab(churn["Int'l Plan"],churn['Churn?'],normalize='index')
ip_ct.plot.bar(stacked=True, ax=ax)

ip_ct2=pd.crosstab(churn["Int'l Plan"],churn['Churn?'])


display(ip_ct,ip_ct2)
plt.savefig("Ä±ntlplan.png")
```

```python
fig, ax = plt.subplots(figsize=(7,7))

vmp_ct=pd.crosstab(churn['VMail Plan'],churn['Churn?'],normalize='index')
vmp_ct.plot.bar(stacked=True, ax=ax)

vmp_ct2=pd.crosstab(churn['VMail Plan'],churn['Churn?'])

display(vmp_ct, vmp_ct2)
plt.savefig("vmailplan.png")
```

```python
fig, ax = plt.subplots(figsize=(20,7))
st_ct=pd.crosstab(churn['State'],churn['Churn?'],normalize='index')
st_ct.plot.bar(stacked=True, ax=ax)
display(st_ct)
plt.savefig("statect.png")
```

```python
fig, ax = plt.subplots(figsize=(10,7))
csc_ct=pd.crosstab(churn['CustServ Calls'],churn['Churn?'],normalize='index')
csc_ct2=pd.crosstab(churn['CustServ Calls'],churn['Churn?'])
csc_ct.plot.bar(stacked=True, ax=ax)
display(csc_ct,csc_ct2)
plt.savefig("custservcall.png")
```

```python
fig, ax = plt.subplots(figsize=(20,7))
sns.countplot(x="State", data=churn, ax=ax, color='Tab:Blue', order=churn['State'].value_counts().index )
plt.savefig("states.png")
```

```python
%matplotlib inline
hist= churn.hist(bins=40,figsize=(12,12))
plt.savefig('hist.png')
```

```python
from sklearn.compose import make_column_transformer
```
